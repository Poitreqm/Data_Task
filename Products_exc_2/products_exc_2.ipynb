````python
import pandas as pd

data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Quantity': [10, 15, 8, 12, 20],
    'Unit Price': [100, 150, 100, 200, 150],
    'City': ['Moscow', 'Saint Petersburg', 'Moscow', 'Yekaterinburg', 'Moscow']
}

df = pd.DataFrame(data)
````
